```python
import pandas as pd
```

```python
# Load data from CSV
print("---- Loading CSV File ----")
csv_data = pd.read_csv("diabetes.csv")
print(csv_data.head(), "\n")
```

```python
# Load data from JSON
print("---- Loading JSON File ----")
json_data = pd.read_json("diabetes.json") 
print(json_data.head(), "\n")
```

```python
# Load data from XLSX
print("---- Loading Excel File ----")
excel_data = pd.read_excel("diabetes.xlsx", sheet_name="Worksheet") 
print(excel_data.head(), "\n")
```

```python
# Load data from API using : 
# 1. pandas.read_json
# 2. pd.dataFrame

print("---- Loading Data from Online API ----")
url = "https://jsonplaceholder.typicode.com/users"
api_data = pd.read_json(url)
print(api_data.head(), "\n")
```
