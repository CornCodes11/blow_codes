```python
# DATA CLEANING WITH PANDAS
import pandas as pd
import numpy as np

# Sample dataset with issues
data = {
    "Student_ID": [101, 102, 103, 104, 105, 105, 105],
    "Name": ["Shardul", "Rahul", "Shyam", "Nehal", "Ravina", "Rakesh" , "Samarth"],
    "Age": [20, np.nan, 22, 21, 200, 21, np.nan],   # Missing + Outlier (200 years!)
    "Marks": [85, 90, np.nan, 88, 95, 95, 100],  # Missing values
    "Joining_Date": ["2023-06-01", "2023-06-05", "06/10/2023", "2023/06/15", "15-06-2023", "2023-06-15", "2025-01-31"]
}
```

```python
df = pd.DataFrame(data)
print("---- Original Data ----")
print(df, "\n")
```

```python
# 1. Handling missing data
# df["Age"].fillna(df["Age"].mean(), inplace=True)
df.fillna({"Age": df["Age"].mean()}, inplace=True)
print(df, "\n")
# Filled age wuth mean
```

```python
# 2. Remove duplicates
print("---- Removing Duplicates ----")
df = df.drop_duplicates()
print(df, "\n")
```

```python
# 3. Fixing data ttypes
print("---- Fixing Data Types ----")

# Convert Joining_Date to datetime
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], dayfirst=True, errors="coerce")
```

```python
# ensuring age is int
df["Age"] = df["Age"].astype(int)

print(df.dtypes, "\n")
print(df, "\n")
```

```python
# 4. treating with outliers
print("---- Handling Outliers ----")

# Using IQR (Interquartile Range) method
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound, "Upper Bound:", upper_bound)

# Treat outliers: replace with median
df.loc[df["Age"] > upper_bound, "Age"] = df["Age"].median()

print("\nFinal Cleaned Dataset:")
print(df)
```
