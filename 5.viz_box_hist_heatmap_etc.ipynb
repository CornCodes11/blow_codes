```python
plt.figure(figsize=(8,6))
plt.hist(df['Glucose'], bins=50, color='pink', edgecolor='red')
plt.title("Histogram of Glucose", fontsize=46)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot histograms for all numeric columns
df.hist(bins=15, figsize=(15,12), color='crimson', edgecolor='pink')
plt.suptitle("Histograms of Diabetes Dataset Features", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=df['Glucose'], color='purple')
plt.title("Boxplot of Glucose", fontsize=26)
plt.ylabel("Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=df, palette="Pastel1")
plt.title("Boxplots of All Features", fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df, palette='coolwarm')
plt.title("Scatterplot: Glucose vs BMI by Outcome", fontsize=16)
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.show()
```

```python
corr_matrix = df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes2.csv')

sns.pairplot(df, hue='Outcome', diag_kind='hist', palette='coolwarm')
plt.suptitle("Pairplot of Features by Outcome", y=1.02, fontsize=16)
plt.show()
```
